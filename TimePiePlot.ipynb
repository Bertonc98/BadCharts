```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime
import scipy.stats as st
import statsmodels.api as sm
import numpy as np
```

```python
data = pd.read_csv('temp.csv', sep=';', decimal=',')
anni=data['Anno'].unique()
month=['Gen','Feb','Mar','Apr','Mag','Giu','Lug','Ago','Set','Ott','Nov','Dic']
pos=1
fig=plt.figure(figsize=(15,15))
for i in anni:
    axcur=fig.add_subplot(3,3,pos)
    plt.title(i)
    tmp = data[data['Anno']==i]
    axcur.pie(tmp['Media'], autopct='%.2f')
    axcur.legend(month, loc=4, bbox_to_anchor=(0, 0))
    pos=pos+1

plt.show()
```

```python
import matplotlib.colors as colors
import matplotlib.cm as cmx


NUM_COLORS = 12

cm = plt.get_cmap('gist_rainbow')
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(111)
ax.set_prop_cycle('color',[cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])

for i in range(1,13):
    tmp=data[data['Mese']==i]
    plt.plot(tmp["Media"], '-o')
plt.cm.colors
plt.legend(month, loc=5, bbox_to_anchor=(1.2, 0.5))
plt.show()
```
