```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime
import scipy.stats as st
import statsmodels.api as sm
import numpy as np
```

```python
data = pd.read_csv('temp.csv', sep=';', decimal=',')
anni=data['Anno'].unique()
month=['Gen','Feb','Mar','Apr','Mag','Giu','Lug','Ago','Set','Ott','Nov','Dic']


pos=1

NUM_COLORS = 12

cm = plt.get_cmap('gist_rainbow')
fig = plt.figure(figsize=(14,7))


for i in anni:
    axcur = fig.add_subplot(3,3,pos)
    axcur.set_prop_cycle('color',[cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])
    plt.title(i)
    tmp = data[data['Anno']==i]
    axcur.pie(tmp['Media'], autopct='%.2f', pctdistance=1.1)
    pos=pos+1
    
plt.legend(month, loc=4, bbox_to_anchor=(0, 0))
plt.show()
```
