```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime
import scipy.stats as st
import statsmodels.api as sm
import numpy as np
```

```python
data = pd.read_csv('csv/temp.csv', sep=';', decimal=',')
anni=data['Anno'].unique()
month=['Gen','Feb','Mar','Apr','Mag','Giu','Lug','Ago','Set','Ott','Nov','Dic']


pos=1

NUM_COLORS = 12

cm = plt.get_cmap('summer')
fig = plt.figure(figsize=(15,15))


for i in anni:
    axcur = fig.add_subplot(3,3,pos)
    axcur.set_prop_cycle('color',[cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])
    plt.title(i)
    tmp = data[data['Anno']==i]
    axcur.pie(tmp['Media'], autopct='%.2f', pctdistance=1.2)
    pos=pos+1
    if pos==8:
        plt.legend(month, loc=9, bbox_to_anchor=(2.5, 0.65), ncol=6, fontsize='x-large')
    
print(pos)
plt.show()
```

```python
import matplotlib.colors as colors
import matplotlib.cm as cmx


NUM_COLORS = 12

cm = plt.get_cmap('gist_rainbow')
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(111)
ax.set_prop_cycle('color',[cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])

for i in range(1,13):
    tmp=data[data['Mese']==i]
    plt.plot([1,2,3,4,5,6,7], tmp["Media"], '-o')

plt.legend(month, loc=5, bbox_to_anchor=(1.12, 0.5), fontsize='x-large')
plt.show()
```

```python
fig = plt.figure(figsize=(14,7))
cm = plt.get_cmap('summer')
plt.title("Tempertures")
ax = fig.add_subplot(111)
tmp = data[data['Anno']==2014]
ax.set_prop_cycle('color',[cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])
plt.pie(tmp['Media'], autopct='%.2f', pctdistance=1.1, wedgeprops={"edgecolor":"k",'linewidth': 1, 'linestyle': 'solid', 'antialiased': True})
plt.legend(month, loc=9, bbox_to_anchor=(1.3, 0.66), ncol=2, fontsize='x-large')
plt.show()
```
