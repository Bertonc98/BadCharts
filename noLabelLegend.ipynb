```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

spesa = pd.read_csv('csv/spesa.csv', sep=';', index_col=0, engine='python', decimal=",")

spesa=spesa[:22]
```

```python
beniAlimentari = spesa[(spesa["Tipologia di beni"] == 'Beni alimentari')]
y=list(np.arange(1,10,1))
plt.barh(y, beniAlimentari["Spesa media mensile familiare (in euro)"], color=(0.996078431372549, 0.7686274509803922, 0.30980392156862746))
plt.show()
```

```python
beniAlimentari = spesa[(spesa["Tipologia di beni"] == 'Beni alimentari')]
labels=list(beniAlimentari.Categoria)
plt.xlabel("Spesa media mensile familiare (in euro) per i beni alimentari")
plt.barh(labels,beniAlimentari["Spesa media mensile familiare (in euro)"], color=(0.996078431372549, 0.7686274509803922, 0.30980392156862746))
plt.show()
```
