```python
import matplotlib.pyplot as plt
import numpy as np

circle1 = plt.Circle((0.3, 0.1), 0.1, color='r')
circle2 = plt.Circle((0.6, 0.2), 0.2, color='blue')
circle3 = plt.Circle((1.3, 0.5), 0.5, color='g')

fig, ax = plt.subplots(figsize=(15,10), ) 
plt.axis("Equal")
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)

plt.xticks(np.arange(0.0,2.1,0.1), ["","","","Red","","","Blue""","","","","","","","Green"])
plt.yticks(np.arange(0.0,1.1,0.1))
plt.show()
```
