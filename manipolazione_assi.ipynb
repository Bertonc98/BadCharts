```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

heroes = pd.read_csv('csv/heroes.csv', sep=';', index_col=0)

#confrontare numero eroi marvel e dc manipolando gli assi

heroes[:10]
```

```python
#seleziono tutti i supereroi della marvel
marvel_heroes=heroes.loc[heroes['Publisher']=='Marvel Comics']

#seleziono tutti i supereroi della dc
dc_heroes=heroes.loc[heroes['Publisher']=='DC Comics']
```

```python
#conto quanti eroi della marvel ci sono
number_marvel_heroes = len(marvel_heroes.index)

#conto quanti eroi della dc ci sono
number_dc_heroes = len(dc_heroes.index)
(number_marvel_heroes , number_dc_heroes)
```

```python
plt.figure(figsize=(15,5))
#grafico sbagliato
plt.subplot(121)
x = ['Marvel Comics', 'DC Comics']
y = [number_marvel_heroes, number_dc_heroes]
plt.bar(x, y, color=(0.9,0.5,0.4))
plt.ylabel('Numero Supereroi')
plt.ylim(150, 400)

#grafico giusto
plt.subplot(122)
x = ['Marvel Comics', 'DC Comics']
y = [number_marvel_heroes, number_dc_heroes]
plt.bar(x,y,  color=(0.403921568627451, 0.6627450980392157, 0.8117647058823529))
plt.ylabel('Numero Supereroi')
plt.show()
plt.show()
```
