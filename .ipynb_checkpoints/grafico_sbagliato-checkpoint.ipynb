```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('csv/immatricolati-anno-accademico-2018-19.csv', sep=';', index_col=0, engine='python', decimal=",")
df[:10]
```

```python
#GRAFICO SBAGLIATO PER LA TIPOLOGIA DEL DATO
```

```python
#il numero di studenti immatricolati è di tipo discreto
iulm = df[(df.AteneoNOME == 'Milano IULM')]
colors = (   (1.0, 1.0, 0.7490196078431373), 
             (0.6313725490196078, 0.8509803921568627, 0.6078431372549019),
             (0.9137254901960784, 0.6392156862745098, 0.788235294117647),
             (0.8470588235294118, 0.7019607843137254, 0.396078431372549),
             (0.6509803921568628, 0.7411764705882353, 0.8588235294117647)
         )
plt.pie(iulm.Imm, labels=list(iulm.CorsoNOME), colors=colors)
plt.title("Numero di studenti immatricolati alla IULM nell'a.a. 2018-2019")
plt.show()
```

```python
iulm = df[(df.AteneoNOME == 'Milano IULM')]
labels=list(iulm.CorsoNOME)
plt.barh(labels,iulm.Imm, color= (0.10980392156862745, 0.5647058823529412, 0.6))
plt.xlabel("Numero studenti immatricolati")
plt.show()
```
