```python
import matplotlib.pyplot as plt
import random as rand
```

```python
plt.figure(figsize=(15, 5))
list = []
for i in range(0, 5):
    tmp = []
    for n in range(0, 30):
        tmp.append(rand.randint(1, 100))
    list.append(tmp)
    plt.plot(list[i])
plt.show()
```

```python
col = ['gray', 'red', 'gray', 'green', 'gray']
plt.figure(figsize=(15, 5))
for i in range(0, len(list)):
    if i%2 != 0:
        plt.plot(list[i], color=col[i])
    else:
        plt.plot(list[i], color=col[i], alpha=.5)
plt.show()
```
