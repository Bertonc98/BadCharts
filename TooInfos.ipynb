```python
import matplotlib.pyplot as plt
import random as rand
```

```python
plt.figure(figsize=(15, 5))
lista = []
for i in range(0, 5):
    tmp = []
    for n in range(0, 30):
        tmp.append(rand.randint(1, 100))
    lista.append(tmp)
    plt.plot(lista[i])
plt.show()
```

```python
col = ['gray', 'red', 'gray', 'green', 'gray']
plt.figure(figsize=(15, 5))
for i in range(0, len(lista)):
    if i%2 != 0:
        plt.plot(lista[i], color=col[i])
    else:
        plt.plot(lista[i], color=col[i], alpha=.5)
plt.show()
```

```python
col = ['blue', 'orange', 'green', 'red', 'purple']
plt.figure(figsize=(17, 3))
plt.subplot(151)
pos=1
for i in range(0, len(lista)):
    if i==0:
        ax=plt.subplot(1,5,pos)
    else:
        plt.subplot(1,5,pos, sharey=ax)
    plt.plot(lista[i], color=col[i])
    
    pos=pos+1
plt.show()
```
